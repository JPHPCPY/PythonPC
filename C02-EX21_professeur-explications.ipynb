La première étape consiste à importer :

* la bibliothèque <span style='color:red' > matplotlib.pyplot </span> avec l'alias *plt*. Elle sert à réaliser le graphique ;
* la bibliothèque <span style='color:red' > numpy </span> . Elle fournit de nombreux outils mathématiques.

```python
import matplotlib.pyplot as plt
import numpy as np
```

On définit deux listes : 
* _x_ qui rassemble les concentration $C_i$ en chlorure de calcium apporté en $\mathrm{mmol\cdot L^{-1}}$ ; 
* _y_ qui rassemble les conductances $G$ en mS.

Dans l'exercice, il est demandé à l'élève de compléter ces deux listes.

```python
# Données expérimentales
x = [10.0, 7.5, 5.0, 2.5, 1.0] # Ci en mmol/L
y = [5.88, 4.41, 2.94, 1.47, 0.59] # G en mS
```

On crée le graphique à l'aide de la bibliothèque <span style='color:red' > matplotlib.pyplot </span>

* ``plt.title()`` permet d'ajouter un titre ; 
* ``plt.xlabel()`` permet d'ajouter un titre à l'axe des abscisses ; 
* ``plt.ylabel()`` permet d'ajouter un titre à l'axe des ordonnées ; 
* ``plt.plot(x,y)`` permet de tracer les courbes.
* ``plt.grid()`` permet d'ajouter une grille au graphique ;
* ``plt.grid`` permet d'ajouter une grille ;
* ``plt.xticks`` et ``plt.yticks`` permettent de gérer les graduations de chaque axe. 


On réalise ensuite une régression linéaire à l'aide de la fonction **polyfit** de la bibliothèque <span style='color:red' > matplotlib.pyplot </span>. _modele_ est une suite de deux termes où le premier ``modele[0]`` est le coefficient directeur de la droite. Le second terme est ``modele[1]`` est l'ordonnée à l'origine.

On crée un liste _xm_ allant de 0 à 10 avec 10 valeurs  l'aide de la commande ``np.linspace(0, 10, 10)``.
On crée également une liste _ym_ rassemblant les images modélisées de chaque valeur de _xm_.

On trace alors la droite de régression à l'aide de la commande ``plt.plot``.

* ``plt.legend()`` permet d'afficher la légende dans le graphique.

Enfin,  ``plt.show()`` permet d'afficher le graphique ainsi obtenu.

```python
# Affichage
plt.title("Courbe d\'étalonnage G=f(C)")
plt.xlabel("C (mmol/L)")
plt.ylabel("G (mS)")
plt.plot(x, y, "bo",label="points expérimentaux")
plt.axis(xmin=0,xmax=10,ymin=0, ymax=6)
plt.grid(linestyle="-.")
plt.xticks(range(11))
# Modelisation fonction linéaire
modele=np.polyfit(x,y,1)
xm = np.linspace(0, 10, 10)
ym = modele[0]*xm + modele[1]
plt.plot(xm, ym, "r-", label="modèle y="+str(round(modele[0],3))+r"$\times x_m + $"+str(round(modele[1],3)))
print("Coefficient directeur du modèle " + str(round(modele[0],3)))
print("Ordonnée à l'origine du modèle " + str(round(modele[1],3)))
plt.legend()
plt.show()
```
